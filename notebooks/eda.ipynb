```python
import pandas as pd
import numpy as np
from os import listdir
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
path = "../data/"

train = pd.read_csv(path+"train.csv")

train.set_index("Accident_ID",inplace=True)
print(train.shape)
train.head()
```

```python
train.describe().T
```

```python
train.nunique().nlargest(10)
```

```python
cat_cols = ['Accident_Type_Code','Violations']
num_cols = list( set(train.columns) - set(cat_cols) )
```

```python
counter = 1
plt.figure(figsize=(8,4))
for col in cat_cols:
    if train[col].nunique() < 10:
        plt.subplot(1,2,counter)
        counter += 1
        train [col].value_counts().plot.bar()
        plt.xticks(rotation = 45)
        plt.title(f'{col}\n(skewness {round(train[ col ].skew ( ),2)})')

plt.tight_layout(pad = 1.0, w_pad = 1.0, h_pad = 1.0)
plt.show ( )
```

> The above features cannot be directly for Linear models

Currently 10 features are there if we make both the label encoded features to one hot encoded we will have ` 8 +13 = 21` features.

And we have 10000 entries > 441, hence we can make those values one hot encoded

```python
train[num_cols].hist(bins=50, figsize=(16,8), ec='w')
plt.show()
```

#### Both the plots(above and below) are same but I am just adding them to understand the code better

```python
%%time
counter = 1
plt.figure(figsize=(16,8))
for i in num_cols:
    if train[i].nunique() >= 10:
        plt.subplot(2,4,counter)
        counter += 1
        sns.distplot ( train [ i ] )
        plt.title(f'{i} (skewness {round(train[ i ].skew ( ),1)})')
        plt.xticks(rotation = 45)
plt.tight_layout()
plt.show ( )
```

```python
for feature in ['Total_Safety_Complaints','Adverse_Weather_Metric']:
    print(feature.rjust(30,'-'))
    print(train[feature].skew())
    print(np.log1p(train[feature]).skew())
    print(np.power(0.01,train[feature]).skew())
```

> We can first remove the skewness then apply the scaling

```python
for feature in ['Total_Safety_Complaints','Adverse_Weather_Metric']:
    plt.figure(figsize=(8,4));
    plt.title(f'Normal {round(train[feature].skew(),1)} v/s Log normal plot {round(np.log1p(train[feature]).skew(),2)}')
    sns.distplot(train[feature],color='g',label="Normal")
    sns.distplot(np.log1p(train[feature]),label="Log Normal")
    plt.legend()
plt.show()
```
