## Logistic regression

```python
from sklearn.linear_model import LogisticRegressionCV
from sklearn.metrics import f1_score
```

```python
results = pd.DataFrame(index=["Train","Validation","Test"])
```

```python
def add_results(model,name,metric):
    """
        model: String
        model specs and tuning
    """
    results.loc[name.title(),model] = metric
    print(results.loc[:,model])
    
def show_results():
    print(results.T)
    
    
def calc_score(actual,predicted):
    print(f1_score(actual,predicted,average="weighted"))
```

```python
%%time
lr = LogisticRegressionCV(cv=2,class_weight="balanced", max_iter=7000,
                            n_jobs=-1,verbose=1,random_state=5,
                                penalty="l2",scoring="f1_weighted",solver="saga")

lr.fit(X_train,y_train)
```

```python
lr.score(X_train,y_train)
```

```python
model = "Logistic cv:2 weight:balanced solver:saga"

add_results(model,"Train",39.6917)
```

```python
results.T
```

```python
pred = lr.predict(X_test)
print("Cross val score")
f1_score(y_test,pred,average="weighted")
```

```python
add_results(model,"validation",39.6706)
```

```python
# Leaderboard score
# 37.28671
add_results(model,"test",37.28671)
```

## One hot encoded

```python
from sklearn.compose import ColumnTransformer

from sklearn.preprocessing import OneHotEncoder
```

```python
train.columns
```

```python
cat_cols = ['Accident_Type_Code','Violations']
num_cols = list( set(train.columns) - set(cat_cols) - {target})
```

##### Repeat the above steps with one hot encoded

```python
categorical_transformer = Pipeline(steps=[
    ('onehot', OneHotEncoder(handle_unknown='ignore'))
])

numeric_transformer = Pipeline(steps=[
    ('scaler', StandardScaler())
])


preprocessor = ColumnTransformer(
    transformers=[
        ('cat', categorical_transformer, cat_cols),
        ('num',numeric_transformer, num_cols)
    ]
)
```

```python
one_hot_pipeline = Pipeline([
    ("preprocessor", preprocessor),
    ("Logistic",transform_predict[1:])
])

one_hot_pipeline.fit(X_train,y_train)
```

```python
name = "Logistic - One hot encoded"

add_results(name,"train", one_hot_pipeline.score(X_train,y_train)*100)
add_results(name,"validation", one_hot_pipeline.score(X_test,y_test)*100)
```

```python
add_results(name,"test", 57.79900)
```

## Reducing skewness in features - Total_Safety_Complaints-> Log1p, Adverse_Weather_Metric-> power transform

```python
id(train)
```

```python
train1 = train.copy()
id(train1)
```

```python
np.log1p(train1['Total_Safety_Complaints']).skew()
```

```python
train1['Total_Safety_Complaints'] = np.log1p(train1['Total_Safety_Complaints'])
```

```python
np.power(0.1, train1['Adverse_Weather_Metric']).skew()
```

```python
train1['Adverse_Weather_Metric'] = np.power(0.1, train1['Adverse_Weather_Metric'])
```

```python
sns.boxplot(data=train1.loc[:,"Safety_Score":],orient="h")
```

```python
X = train1.drop(columns=[target])

y = train1[target]
```
